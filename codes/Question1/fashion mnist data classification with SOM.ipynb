```python
import numpy as np
import matplotlib.pyplot as plt
import keras
import tensorflow as tf
import random
```

```python
from keras.datasets import fashion_mnist 
```

```python
(x_train, y_train),(x_test, y_test) = fashion_mnist.load_data()
x_train = x_train[:1000]/255
test_data = x_test[:300]/255
x_train = np.reshape(x_train, (1000,28*28))
test_data = np.reshape(test_data, (300,28*28))
```

```python
x_train.shape[0]
```

```python
num = 15
n = 28*28
m = num*num
w = np.random.random(size = (n, m))
d = np.zeros(m)
a = 0.9
iter = 0
w_old = w.copy()
R = 1
allowed = [i for i in range(m)]
up_weight = []
while(iter < 10):
  for i in range(x_train.shape[0]):
    for j in range(m):
      d[j] = (w[:, j] - x_train[i])@(w[:, j] - x_train[i])
    idx = np.argmin(d)
    for k in range(R + 1):
      for x in range(2*R + 1):
        if((idx + (k*num-R) + x) in allowed):
          w[:, idx + (k*num-R)+x] = w_old[:, idx + (k*num-R)+x] + a*(x_train[i] - w_old[:, idx + (k*num-R)+x])
        if (k != 0):
          if((idx-(k*num-R) + x) in allowed):
            w[:, idx -(k*num-R)+x] = w_old[:, idx-(k*num-R)+x] + a*(x_train[i] - w_old[:, idx -(k*num-R)+x])
    w_old = w.copy()
  a = 0.7*a
  iter = iter + 1
  if ((iter == 1) or (iter == 2) or (iter == 5) or (iter == 9)):
    up_weight.append(w_old)
```

```python
f, axis = plt.subplots(15,15, figsize=(10,10))
for j in range(m):
  axis[j//15, j%15].imshow(np.reshape(up_weight[0][:,j], (28, 28)), cmap = "Greys")
```

```python
f, axis = plt.subplots(15,15, figsize=(10,10))
for j in range(m):
  axis[j//15, j%15].imshow(np.reshape(up_weight[1][:,j], (28, 28)), cmap = "Greys")
```
